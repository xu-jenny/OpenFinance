```python
import pandas as pd
df = pd.read_csv("co.csv")
df.head()
```

```python
df = df.rename(columns={"Company Name": "companyName", "Layoff Total": "numAffected", "WARN Date": "noticeDate", "Layoff Date(s)": "layoffDate"})
```

```python
def is_standard_date(date):
    try:
        # Try to convert the string to a datetime object
        pd.to_datetime(date)
        return True
    except ValueError:
        # If it fails, it's not a standard date
        return False

# Create a boolean mask where each value is True if the date is not standard and False otherwise
mask = ~df['noticeDate'].apply(is_standard_date)
df[mask]
```
